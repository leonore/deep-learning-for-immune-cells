```
def plot_tsne(X, y, labels=["Unstimulated", "OVA", "ConA"]):
    targets = range(len(labels))
    palette = np.array(sns.color_palette("hls", len(labels)))

    y = np.array(y)

    fig = plt.figure(figsize=(20,20))
    ax = plt.subplot()

    for target, color, label in zip(targets, palette, labels):
        plt.scatter(X[y==target, 0], X[y==target, 1], c=[color], label=label)

    sorted_x = X[X[:,0].argsort()]
    sorted_y = y[X[:,0].argsort()]
    indices = np.array([x for x in range(len(y))])
    sorted_indices = indices[X[:,0].argsort()]

    idx = 0
    while idx < len(sorted_x)-1:
      if sorted_y[idx] != sorted_y[idx+1]:
        plt.annotate(sorted_indices[idx], xy=(sorted_x[idx, 0], sorted_x[idx, 1]), xytext=(sorted_x[idx, 0]+2, sorted_x[idx, 1]+1*np.random.randint(-2, 5)), fontsize=10, arrowprops=dict(arrowstyle='->', lw=0.9, alpha=0.7),bbox=dict(pad=100, facecolor="none", edgecolor="none"))
      idx += 16
    
    #for i, txt in enumerate(y):
    #  plt.annotate(i, (X[i, 0], X[i, 1]), fontsize=8, alpha=0.5) 
    ax.axis('off')
    ax.grid(False)

    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
